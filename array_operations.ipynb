1.Arithmetic Operation

```python
a1=np.array([1,2,3,4,5])
a2=np.array([6,7,8,9,10])
```

```python
#add corresponding element
arr1=a1+a2 
arr1                   #the size of both array need to be same
```

```python
arr2=a1*a2
arr2
```

Deep and Shallow copy

```python
a=np.arange(1,21)
a
```

```python
slice=a[:5]
slice=slice*10
slice                                               #shallow copy
```

```python
a
```

```python
b=a             #deep copy
b[0]=99
a
```

```python
c=a.copy()      #shallow copy
c[0]=8
a
```

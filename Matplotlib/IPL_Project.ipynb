### Loading the Libraries and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

df=pd.read_csv('ipl.csv')
df.head()
```

```python
df.isnull().sum()
```

Now, Here comes some Basic Questions

1.Which team won the most matches?

```python
wins=df["match_winner"].value_counts()
wins
```

```python
sns.barplot(x=wins.values,y=wins.index,palette='viridis')
plt.title("Match wins by a team")
```

2.Toss Decision Trends

```python
toss_decision=df['toss_decision'].value_counts()
toss_decision
```

```python
sns.countplot(x=df['toss_decision'])
```

```python
plt.pie(x=toss_decision.values,labels=toss_decision.index,autopct='%1.1f%%')
plt.title("Toss Decisions")
```

Toss Winner V/S Match Winner

```python
count=df[df['toss_winner']==df['match_winner']]['match_winner'].count()
percentage=(count*100)/df.shape[0]
percentage.round(2)
```

How do team wins?(Runs vs Wickets)

```python
c=df[df['won_by']=='Wickets']['won_by'].count()
sns.countplot(x=df['won_by'])
plt.title('Won By')
```

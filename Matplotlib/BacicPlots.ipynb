MATPLOTLIB

Basic plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,9,10)
y=x+1
```

```python
#plot create line graph
plt.title('First Plot')
plt.xlabel('X')
plt.ylabel('X**2')
plt.plot(x,y)
```

Subplot in matplotlib

```python
#plt.subplot(rows,cols)
#plt.subplot(2,2,1) is used to select the first plot and (2,2,3) select 3 plot
plt.subplot(2,2,1)
y=x**-1
plt.plot(x,y)
plt.subplot(2,2,2)
y=x
plt.plot(x,y)
plt.subplot(2,2,3)
y=x**2
plt.plot(x,y)
plt.subplot(2,2,4)
y=x**3
plt.plot(x,y)
```

Object Oriented Way

Matplotlib allows the aspect ratio,DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword argument
figsize is a tupule of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.2,0.7,0.7])
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.4,0.3,0.3])
axis2.plot(x,y)
```

Types of Plot

```python
#Scatter Plot
plt.scatter(x,y)
```

```python
# histogram
from random import sample
data=sample(range(1,1000),100)
plt.hist(data)
```

Saving plots

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicPlot.png")
```

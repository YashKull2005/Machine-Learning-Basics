SERIES

Series is a one dimensional labeled array capable of holding any datatype.The axis labels are collectively called index.

Creating Series


```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
myList=[10,20,30]
arr=np.array(myList)
d={1:10,2:20,3:30}
```

```python
#Creating series using list
pd.Series(myList,index=labels)
```

```python
#Creating series from array
pd.Series(arr,index=labels)
```

```python
#Creating series from dictionary   keys->labels  and  value->data
pd.Series(d)
```

DATA FRAMES

Data frames are collection of series.

Create a DataFrame

```python
data={
    'Name':['John','Anna','Peter','Linda'],
    'Age':[28,34,29,42],
    'City':['New York','Paris','Berlin','London'],
    'Salary':[65000,75000,80000,60000]
}
df=pd.DataFrame(data,index=[1,2,3,"hello"])  #index used to name row
df
```

```python
data=[
    ['John',28,'New York',65000],
    ['Anna',34,'Paris',70000],
    ['Peter',29,'Berlin',88000],
    ['Linda',42,'London',60000]
]
df2=pd.DataFrame(data)
```

```python
df2
```

```python
columns=["Name","Age","City","Salary"]
df2=pd.DataFrame(data,columns=columns)
df2
```

Selection and Indexing of Columns

```python
df2
```

```python
#index does not work but the column name work
df2['Name']
```

```python
df2['Name'][1]
```

Creating a new column

```python
df2['Designation']=['Doctor','Engineer','Lawyer','Cricketer']
df2
```

```python
#dropping an column
df2.drop(['City','Designation'],axis=1)
df2
```

```python
#droping an column using inplace
#inplace ensure that the change is made in main table while if inplace is not true then it will give us an idea how the table will look if the particular column is droped
df2.drop('Designation',axis=1,inplace=True)
df2
```

```python
df2.drop(['Age','City','Salary'],axis=1)
```

```python
df2.drop(0)   #droping a row
```

```python
df2.loc[[0,1]]
```

```python
#Selecting index location iloc in this we pass row index 
df2.iloc[2]   
```

Selecting Subset of Rows and Columns

```python
df2.loc[[1,2]][['Name','Age']]
```

Conditional Selection

```python
#select those whose age is >30
df2[(df2['Age']>30) & (df2['City']=='Paris')]
```

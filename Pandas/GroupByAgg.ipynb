```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#Group by category and find the sum of sales
x=df.groupby('Category')
```

```python
for k,v in x:
    print(x,v)
```

```python
x['Sales'].sum()
```

```python
#Group by Store and find the sum of sales
y=df.groupby('Store')
y['Sales'].sum()
```

```python
#Group by multiple columns
#Group by category and Store
z=df.groupby(['Store','Category'])
z['Sales'].sum()
```

Aggregation

Basic aggregation function are mean,median,standard deviation(std),variance,min,max,sum,count

```python
df['Sales'].mean()
```

```python
#agg function
df['Sales'].agg(['mean','max','median','min'])
```

```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
#find null values
df.isna()
```

```python
#To find number of na values in a column
df.isna().sum()
```

```python
#To find is every column have a null value or some column is free from it
df.isna().any()
```

Removing missing data (works on basis of rows)

```python
#Drop all those rows that have atleast one missing value
df.dropna()
```

```python
#using Threshold
#remove rows that have na values greater than equal to 3
df.dropna(thresh=3)
```

Filling Missing Values

```python
#fill all the na values with 0 take place column wise
df.fillna(0)
```

```python
#fill missing values with different entries in different column
values={'A':0,'B':1,'C':3,'D':4}
df.fillna(value=values)
```

```python
#fill na with the mean value of column
df.fillna(df.mean())
```

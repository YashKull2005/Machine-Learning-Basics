```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
np.shape(data)
```

```python
#Find the average age of students.
age=data[:,0]
averageAge=age.sum()/5
averageAge=np.mean(data[:,0])
print("Average Age of student is :: ",averageAge)
```

```python
#Extract Math marks of all students.
mathMarks=data[:,1]
mathMarks
```

```python
#Find the highest Science mark.
ScienceMarks=data[:,2]
np.max(ScienceMarks)
```

```python
#Get details of the student who scored more than 90 in Math.
# mathMarks=data[:,1]>1
# for i in range(len(mathMarks)):
#     if(mathMarks[i]==90):
#         print("Student Details :: ",data[i])
mathMarks=data[:,1]>90
data[mathMarks]
```

```python
#Increase Math marks of all students by 5.
data[:,1]=data[:,1]+5
data

```

```python
#Find how many students are younger than 19.
boolean_array=data[:,0]<19
len(data[boolean_array])
```

```python
#Calculate the average marks in each subject (column-wise mean).
columnMean=data[:,1:].mean(axis=0)
columnMean
```

```python
#Get data of students who scored at least 80 in both subjects.
data[(data[:,1]>=80) & (data[:,2]>=80)]
```

```python
#Replace all Science marks < 75 with 0.
data[:,2][data[:,2]<75]=0
data
```
